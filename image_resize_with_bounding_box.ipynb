```python
# importing stuffs 
import cv2
import matplotlib.pyplot as plt 
import numpy as np 
import re
from glob import glob
```

```python
# for testing the box resizing logic
def drawBox(boxes, image):
    for i in range (0, len(boxes)):
        cv2.rectangle(img=image, pt1=(boxes[i][0],boxes[i][1]), pt2=(boxes[i][2],boxes[i][3]), color=(0,0,255))
    cv2.imshow("img",image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
all_files = glob(r'C:\Users\pui_s\Downloads\HouseAddress\images\*.*')
# take only those without xml in the name
for file in all_files:
    # skip all the xml files
    if 'xml' in file:
        continue

    original_image = cv2.imread(file, 3)
    # print(imageToPredict.shape)
    y_, x_, _ = original_image.shape

    new_size = (640, 640)
    x_scale = new_size[1]/x_
    y_scale = new_size[0]/y_
    # print(x_scale,y_scale)
    resized_image = cv2.resize(src=original_image, dsize=new_size);


    # reading box(es) from xml file, and shrink it
    try:
        with open(file.split('.')[0] + r'.xml', 'r') as f:
            raw_xml = f.readlines()
            image_xml = '='.join(raw_xml)
        f.close()

        xmin = [int(int(x)*x_scale) for x in re.findall(r'(?<=<xmin>)(.*)(?=</xmin>)', image_xml)]
        xmax = [int(int(x)*x_scale) for x in re.findall(r'(?<=<xmax>)(.*)(?=</xmax>)', image_xml)]
        ymin = [int(int(y)*y_scale) for y in re.findall(r'(?<=<ymin>)(.*)(?=</ymin>)', image_xml)]
        ymax = [int(int(y)*y_scale) for y in re.findall(r'(?<=<ymax>)(.*)(?=</ymax>)', image_xml)]


        # drawBox([],resized_image)
    #     drawBox([[i,j,k,l] for i,j,k,l in zip(xmin, ymin, xmax, ymax)], resized_image)

        # updating xml 
        for i in range(len(raw_xml)):
            line = raw_xml[i]
            if re.findall(r'xmin', line):
                raw_xml[i] = re.sub(r'(?<=<xmin>)(.*)(?=</xmin>)', str(xmin.pop(0)), line)
            elif re.findall(r'xmax', line):
                raw_xml[i] = re.sub(r'(?<=<xmax>)(.*)(?=</xmax>)', str(xmax.pop(0)), line)
            elif re.findall(r'ymin', line):
                raw_xml[i] = re.sub(r'(?<=<ymin>)(.*)(?=</ymin>)', str(ymin.pop(0)), line)
            elif re.findall(r'ymax', line):
                raw_xml[i] = re.sub(r'(?<=<ymax>)(.*)(?=</ymax>)', str(ymax.pop(0)), line)

        # saving images
        cv2.imwrite(r'C:\Users\pui_s\Downloads\HouseAddress\images_resize\\' + file.split('\\')[-1], resized_image)
        # saving the xml file
        print('saving: ' + r'C:\Users\pui_s\Downloads\HouseAddress\images_resize\\' + file.split('\\')[-1])
        out_f = open(r'C:\Users\pui_s\Downloads\HouseAddress\images_resize\\' + file.split('\\')[-1].split('.')[0] + '.xml', "w")
        out_f.write(''.join(raw_xml))
        out_f.close()
    except:
        None
```
